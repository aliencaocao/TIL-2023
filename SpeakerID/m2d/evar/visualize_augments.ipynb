```python
import pickle

# read the files before_aug.pkl and after_aug.pkl
with open('before_aug.pkl', 'rb') as f:
    before_aug = pickle.load(f)

with open('after_aug.pkl', 'rb') as f:
    after_aug = pickle.load(f)

with open('after_aug_padding.pkl', 'rb') as f:
    after_aug_padding = pickle.load(f)
```

```python
# visualize before aug, after aug, and after aug + padding 
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
plt.plot(before_aug)
plt.title('before aug')
plt.subplot(1, 3, 2)
plt.plot(after_aug)
plt.title('after aug')
plt.subplot(1, 3, 3)
plt.plot(after_aug_padding)
plt.title('after aug + padding')
```
