```python
from ultralytics import YOLO

model = YOLO("yolov8x.pt")
```

```python
model.train(
    data="til.yml",
    epochs=300,
    batch=16,
    nbs=64, # effective batch size
    imgsz=640,
    save=True,
    save_period=1,
    cache="ram",
    device=0,
    workers=0, # often hangs if > 0
    project="D:\\TIL YOLOv8\\yolov8x",
    pretrained=True,
    optimizer="SGD",
    cos_lr=True,
    lr0=1e-5,
)
```
