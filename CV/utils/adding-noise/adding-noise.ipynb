```python
from PIL import Image
import numpy as np
from IPython.display import display
```

```python
img = Image.open("image_0632.png")
img_array = np.array(img)
```

```python
np.random.normal(loc=0, scale=1.0, size=img_array.shape) * 255
```

```python
noise_strength = 25

noise = noise_strength * np.random.normal(loc=0, scale=1.0, size=img_array.shape)
img_array_noisy = np.rint(img_array + noise)
img_array_noisy_clipped = np.clip(img_array_noisy, 0, 255)

img_noisy = Image.fromarray(img_array_noisy_clipped.astype(np.uint8))
display(img_noisy)

img_noisy.save(f"image_0632_noisy_{noise_strength}.png")
```
