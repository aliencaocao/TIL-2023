```python
import os
```

```python
checksum_dict = {}
duplicates_list = []
for plushie_identity in os.listdir("../RT-DETR/dataset/suspects_hand_label"):
    for filename in os.listdir(f"../RT-DETR/dataset/suspects_hand_label/{plushie_identity}"):
        filepath = os.path.join(f"../RT-DETR/dataset/suspects_hand_label/{plushie_identity}", filename)
        with open(filepath, "rb") as f:
            checksum = hash(f.read())
        if checksum in checksum_dict.keys():
            print(f"Duplicate found: {filepath}, duplicate of {checksum_dict[checksum]}")
            duplicates_list.append(filepath)
        checksum_dict[checksum] = filepath
```

```python
new_duplicates_list = [f"../RT-DETR/dataset/suspects_hand_label_unique/{filepath[20:]}" for filepath in duplicates_list]
new_duplicates_list
```

```python
# delete each item in new_duplicates_list
for filepath in new_duplicates_list:
    os.remove(filepath)
```
