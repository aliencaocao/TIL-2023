```python
import pandas as pd
```

```python
df = pd.read_csv('IMDA_metadata_number_audios.csv')
df
```

```python
def new_filepath(x):
    return x[16:]

df['new_filepath'] = df['path'].apply(new_filepath)
df
```

```python
import shutil
import os

counter = 0

def copy_file(src_path, dest_path):
    global counter
    # Create dest folder if it doesn't exist
    os.makedirs(os.path.dirname(dest_path), exist_ok=True)
    print(f'{counter}: Copying {src_path} to {dest_path}')
    shutil.copy2(src_path, dest_path)
    counter += 1

def process(row):
    copy_file(row['path'], os.path.join('imda_truncated', row['new_filepath']))

df.apply(process, axis=1)
```
